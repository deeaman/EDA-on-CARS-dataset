## Before data cleaning

```python
print("Number of Missing Values in our data set\n")
missing_df = data.isnull().sum().to_frame().reset_index().rename({"index" : 'Variable', 0: 'Missing Values'}, axis =1)
display(missing_df.style.background_gradient('gnuplot2_r'))
print("\n Percentage of Missing Values in our data set")
display((data.isnull().sum() / (len(data.index)) * 100).head(20).to_frame().rename({0:'Count'}, axis = 1).style.background_gradient('gnuplot2_r'))
round((data.isnull().sum() / (len(data.index)) * 100) , 2).plot(kind = 'barh',color ='#bf0606')

plt.title("Percentage of Missing values");
```

# Data cleaning

```python
data['market_category'].fillna(data['market_category'].mode()[0], inplace = True)
data['engine_fuel_type'].fillna(data['engine_fuel_type'].mode()[0], inplace = True)
   
```

```python
data.isnull().sum()
```

```python
data['engine_hp'].fillna(data['engine_hp'].mean(), inplace = True)
```

```python
data.isnull().sum()
```

```python
data['engine_cylinders'].fillna(data['engine_cylinders'].median(), inplace = True)
data['number_of_doors'].fillna(data['number_of_doors'].median(), inplace = True)
```

```python
data.isnull().sum()
```

```python
print("Number of Missing Values in our data set\n")
missing_df = data.isnull().sum().to_frame().reset_index().rename({"index" : 'Variable', 0: 'Missing Values'}, axis =1)
display(missing_df.style.background_gradient('gnuplot2_r'))
print("\n Percentage of Missing Values in our data set")
display((data.isnull().sum() / (len(data.index)) * 100).head(20).to_frame().rename({0:'Count'}, axis = 1).style.background_gradient('gnuplot2_r'))
round((data.isnull().sum() / (len(data.index)) * 100) , 2).plot(kind = 'barh',color ='#bf0606')

plt.title("Percentage of Missing values");
```

# Which companies making expensive cars

```python
plt.figure(figsize=(20,10))
sns.barplot(x='make', y="msrp", data=data,order = data.groupby(by = 'make')['msrp'].mean().sort_values( ascending=False).index )
plt.xticks(rotation = 90)
plt.show()
```

## Observation
#### Buggatti producing most expensive cars
#### followed by Maybach and Rolls-Roys

## Vehicle style vs msrp


```python
sns.set({'figure.figsize':(20,10)})
VS=sns.barplot(x=data['msrp'], y=data['vehicle_style'] , order = data.groupby(by = 'vehicle_style')['msrp'].mean().sort_values( ascending=False).index )
plt.title('Vehicle Style and MSRP', weight='bold', fontsize=18)
plt.xlabel('MSRP', weight='bold',fontsize=14)
plt.ylabel('Vehicle Style', weight='bold', fontsize=14)
plt.xticks(weight='bold')
plt.yticks(weight='bold')
```

## No. of cylinder -> more engine hp -> more msrp

```python
plt.figure(figsize=(15,5))
#first row, first col
ax1 = plt.subplot2grid((1,2),(0,0))
plt.scatter(x=data['engine_cylinders'], y=data['engine_hp'], color='maroon', alpha=0.7)
plt.title('Engine Cylinders on Engine HP', weight='bold', fontsize=18)
plt.xlabel('Engine Cylinders', weight='bold',fontsize=14)
plt.ylabel('Engine HP', weight='bold', fontsize=14)
plt.xticks(weight='bold')
plt.yticks(weight='bold')


#first row sec col
ax1 = plt.subplot2grid((1,2), (0, 1))
sns.regplot(x=data['engine_hp'], y=data['msrp'], color='maroon')
plt.title('Engine HP on Engine MSRP', weight='bold', fontsize=18)
plt.xlabel('MSRP', weight='bold',fontsize=14)
plt.ylabel('Engine HP', weight='bold', fontsize=14)
plt.xticks(weight='bold')
plt.yticks(weight='bold')
```

### Engine_hp -> highway_hp -> MSRP

```python
plt.figure(figsize=(15,5))
#first row, first col
ax1 = plt.subplot2grid((1,2),(0,0))
sns.regplot(x=data["engine_hp"], y=data["highway_mpg"], line_kws={"color":"red","alpha":1,"lw":5})
plt.title('Highway MPG and Engine HP', weight='bold', fontsize=18)
plt.xlabel('Engine HP', weight='bold',fontsize=14)
plt.ylabel('Highway MPG', weight='bold', fontsize=14)
plt.xticks(weight='bold')
plt.yticks(weight='bold')

#first row sec col
ax1 = plt.subplot2grid((1,2), (0, 1))
sns.regplot(x=data["msrp"], y=data["highway_mpg"], line_kws={"color":"red","alpha":1,"lw":5})
plt.title('Highway MPG and MSRP', weight='bold', fontsize=18)
plt.xlabel('MSRP', weight='bold',fontsize=14)
plt.ylabel('Highway MPG', weight='bold', fontsize=14)
plt.xticks(weight='bold')
plt.yticks(weight='bold')
```
